```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid', palette='muted', font_scale=1.1)

# Load
credit_df = pd.read_csv("../data/raw/creditcard.csv")

# Class imbalance
imbalance = credit_df['Class'].value_counts(normalize=True) * 100
print(f"Legitimate: {imbalance[0]:.2f}%, Fraud: {imbalance[1]:.2f}%")

# Correlation with target
plt.figure(figsize=(12,6))
credit_df.corr()['Class'].sort_values().plot(kind='bar', color='skyblue')
plt.title("Correlation with Fraud (Credit Card Data)")
plt.show()

# Fraud vs Amount
plt.figure(figsize=(10,5))
sns.boxplot(x='Class', y='Amount', data=credit_df)
plt.ylim(0, 5000)  # Focus on majority of transactions
plt.title("Transaction Amount by Class")
plt.show()

# Fraud vs Time
plt.figure(figsize=(10,5))
sns.kdeplot(data=credit_df, x='Time', hue='Class', common_norm=False, fill=True)
plt.title("Fraud Density over Time (Seconds since first transaction)")
plt.show()

# Save processed
credit_df.to_csv("../data/processed/credit_cleaned.csv", index=False)
print("Credit card EDA complete.")
```
